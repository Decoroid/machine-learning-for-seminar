```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets, preprocessing
from sklearn import svm
%matplotlib inline
```

```python
iris = datasets.load_iris()
```

```python
X = iris.data[:100, :2]
y = iris.target[:100]
X = preprocessing.minmax_scale(X)
```

```python
clf = svm.SVC(kernel = 'linear')
clf.fit(X, y)
```

```python
w = clf.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(0, 1)
yy = a * xx - clf.intercept_[0] / w[1]
```

```python
plt.axes().set_aspect('equal', 'datalim')
plt.scatter(X[:50, 0], X[:50, 1], marker='o', color='red', s=50, alpha=.5, label='setosa')
plt.scatter(X[50:, 0], X[50:, 1], marker='x', color='blue', s=50, alpha=.5, label='versicolor')
plt.plot(xx, yy, color='green')
plt.legend(loc = 'upper left')
plt.xlabel('sepal length[cm]')
plt.ylabel('petal length[cm]')
plt.grid()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
x_train, x_test, y_train, y_test = train_test_split(X, y.ravel(), test_size=0.3)
```

```python
clf = svm.SVC(kernel = 'linear')
clf.fit(x_train, y_train)
```

```python
predict = clf.predict(x_test)
print('{0}%'.format(accuracy_score(y_test, predict) * 100))
```
