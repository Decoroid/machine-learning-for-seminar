```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets, preprocessing
from sklearn import svm
%matplotlib inline
```

```python
data_num = 500
X, y = datasets.make_moons(data_num, noise=0.2)
```

```python
clf = svm.SVC(kernel = 'rbf', gamma=1000.0, C=1.0)
clf.fit(X, y)
```

```python
h = .02
x_min, x_max = X[:, 0].min(), X[:, 0].max()
y_min, y_max = X[:, 1].min(), X[:, 1].max()
xx, yy = np.meshgrid(np.arange(x_min, x_max+1, h), np.arange(y_min, y_max+1, h))
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
```

```python
Z = Z.reshape(xx.shape)
plt.contour(xx, yy, Z, cmap=plt.cm.Greens)
plt.scatter(X[:, 0], X[:, 1], marker='o', c=y, s=50, alpha=.5, cmap=plt.cm.Spectral)
```
